```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

Plotting linear models:

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"],
          scatter_kws={"s":20})
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",col="day",
          aspect=0.6,size=8)
```
