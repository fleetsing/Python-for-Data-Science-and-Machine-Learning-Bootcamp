```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

Functional method:

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")

plt.show()
```

Object oriented method:

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,"r")
axes2.plot(y,x,"b")

axes1.set_title("Larger Plot")
axes2.set_title("Smaller Plot")

plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,"b")
axes[0].set_title("First")

axes[1].plot(y,x,"r")
axes[1].set_title("Second")

plt.tight_layout()
```

Figure size and DPI:

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(10,4),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

Legends:

```python
fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,x**2,label="X squared")
ax.plot(x,x**3,label="X cubed")

ax.legend()
plt.show()
```

Saving figures:

```python
fig.savefig("my_picture.png")
```

Plot appearance:

```python
fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color="#7427f9",linewidth=0.5,alpha=0.7,linestyle="--",
        marker="o",markersize=8,markerfacecolor="b",markeredgewidth=2,markeredgecolor="r")

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,)

ax.set_xlim([0,1])
ax.set_ylim([0,1])

plt.show()
```
